```python
%matplotlib inline
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
import math

m = 100000
n = [m*0.1, m*0.2, m*0.4, m*0.8]
k = np.arange(0, 100, 0.1)

l_color = ['#fcdfcf','#fabe9e', '#f79e6e','#f36618' ]
labels = ["n=0.1m", "n=0.2m", "n=0.4m", "n=0.8m"]
x = k
y = [[]] * 4
y2 = y
for i in range(4):
    y[i] = np.power(1 - np.power( 1 - 1/m, k*n[i]), k)
    y2[i]= np.power(1 - np.exp(-k*n[i]/m), k)
    plt.figure(1)
    plt.plot(x, y[i], l_color[i], label=labels[i])
    plt.figure(2)
    plt.plot(x, y2[i], l_color[i],label=labels[i])
    
plt.figure(1)
plt.grid()
plt.legend()
plt.title("false positive rate")
plt.xlabel("k")
plt.savefig("false positive rate", dpi=200)

plt.figure(2)
plt.grid()
plt.legend()
plt.xlabel("k")
plt.title("approximate false positive rate")
plt.savefig("approximate false positive rate", dpi=200)

plt.show()
```

```python
p = np.arange(0.01,1,0.01)
gp = [[]] * 4
for i in range(4):
    gp[i] = -m/n[i] * np.log(p) *np.log(1-p)
    plt.figure(3)
    plt.plot(p, gp[i], l_color[i], label=labels[i])
    
plt.grid()
plt.legend()
plt.title("value of g(p)")
plt.xlabel("p")
plt.savefig("value-of-gp", dpi=200)
```

```python
l_color = ['#e67300','#ff8000', '#ff8c1a','#ff9933', '#ffa64d', '#ffb366', '#ffbf80', '#ffcc99' ]
labels = ["log2(m)=8", "log2(m)=12", "log2(m)=16", "log2(m)=20", "log2(m)=24", "log2(m)=28", "log2(m)=32", "log2(m)=36"]
m = np.power(2,[8,12,16,20,24,28,32,36])
n = np.arange(0,16,0.1)
n = np.power(10,n)

x = []
y = [[]] * 8

for i in range(8):
    ax = plt.subplot(111)
    ax.set_xscale("log", nonposx='clip')
    ax.set_yscale("log", nonposy='clip')
    x = []
    for _ in n:
        k = (m[i]/_)*np.log(2)
        y_ = np.power(1 - np.power( 1 - 1/m[i], k*_), k)
        x = np.append(x,_)
        y[i] = np.append(y[i],y_)
    plt.plot(x, y[i], l_color[i], label = labels[i])
plt.grid()
plt.legend()
plt.ylim((1e-8, 1.5))
plt.title("false positive as a function of number of elements n")
plt.xlabel("n")
plt.ylabel("p")
plt.savefig("value-of-f_n", dpi=200)
```
